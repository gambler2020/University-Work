# Cleaning the Data

```python
#now we can clearly see that how much dirty our table is need to clean in order to use it.
#we have "NaN" in many rows and doesn't have any meaning so better to remove all the rows that have "NaN" cells.
```

```python
df1=df.drop(df.index[0:4], inplace=True) #drop() is used to delete particular row. here we need to delete first 4 rows
                                         #we used slicing index to delete rows from 0th to 3rd rows at once.
                                         #inplace is used here to reflect changes in the dataframe.
```

```python
df.drop([5], inplace=True)
```

```python
df.dropna(inplace=True) #dropna is used to delete all the rows that have NaN in the cell.
```

```python
df.head(5)
```

```python
#Data looks almost clean but actually not! now our job is to fix column indexes
#we have numbers in the column index need o rename
```

```python
df1=df.rename(columns={df.columns[0]: "Time Period", 
                       df.columns[1]: "Estimated average % of the population that had COVID-19", 
                       df.columns[2]: "95% Lower credible interval",
                      df.columns[3]: "95% Upper credible interval",
                      df.columns[4]: "Estimated number of people testing positive for COVID-19",
                      df.columns[5]: "95% Lower credible interval",
                      df.columns[6]: "95% Upper credible interval",
                      df.columns[7]: "Estimated average ratio of the population that had COVID-19",
                      df.columns[8]: "95% Lower credible interval",
                      df.columns[9]: "95% Upper credible interval"})
```

```python
df1.head(5)
```

```python
#we have renamed but now it is seen that we have one more rows that have similar as index needs to delete.
```

```python
df2=df1.drop([4])
```

```python
df2.set_index(["Time Period"], inplace=True) #set row index instead of having default row index
```

```python
df2.head(5)
```

```python
#now thinking to use row index as a column in my plotting graph then have to reset to normal as it was before.
```

```python
df2.reset_index(level=["Time Period"], inplace=True)
```

```python
df2.head(5)
```

# Visualization

```python
x = (df2["Estimated number of people testing positive for COVID-19"]) # x is assigned to a column that have +ve cases
y= (df2["Time Period"])                                               # y is assigned to the column Time Period
```

```python
fig, ax = plt.subplots(figsize=(40, 6)) # used for the layout of the figures
plt.setp(ax.get_xticklabels(), rotation = 45) # to rotates text at 45 degree on xaxis for clear looks.
plt.plot(y, x)                              # plotting the line grpah using x and y axis
plt.xlabel("(Time)", fontsize=20)           # Name to represent x-axis
plt.ylabel("(No. of Positive Cases)")       # Name to represent y-axis
plt.title("(Covid Cases in Wales)\nDaily positive cases Covid-19 VS Time graph", color="red") #use to represent the title of the graph
fig.autofmt_xdate() # for right alignment on date formated data type on x axis
plt.savefig("output.jpg")
plt.show() # used for display the graph
```
