# TASK 2

```python
# Choose any five countries of your choice and make a new data frame containing only five countries
# and their GDP data (from 1990 until 2019).
```

```python
df_countries=df[df.columns[~df.columns.isin(['Country Code','Indicator Name', "Indicator Code"])]]
```

```python
df_5_countries=df_countries.drop(df_countries.iloc[:, 1:31], axis=1).sample(5)
```

```python
df_5_countries
```

# TASK 3

```python
# Perform z-score standardisation on the GDP data (1990-2019) of all the five countries.
```

```python
z_score=df_5_countries.drop("Country Name", axis=1)
```

```python
z_score=(z_score-z_score.mean())/z_score.std()
```

```python
z_score
```
